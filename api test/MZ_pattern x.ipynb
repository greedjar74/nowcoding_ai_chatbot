```python
%%time

user_content = '''
2. 명령어 설명
 해당 문제는 지정된 명령어가 존재하는데 각각의 명령어의 역할에 대해서 가르쳐줄게. 명령어는 함수 형태로 이미 만들어져있기에 명령어를 정의할 필요는 없으며 인자를 입력받지도 않아. 이제 명령어에 대해서 설명해줄게

moveForward() : 아바타가 바라보는 방향으로 1칸 앞으로 전진. 단, 앞에 길이 없는데 실행할 경우 오류발생.
turnLeft() : 제자리에서 왼쪽으로 90도 회전
turnRight() : 제자리에서 오른쪽으로 90도 회전
turnBack() : 제자리에서 180도 회전
pathAhead() : 앞에 길이 있는지 반환. 가능하면 True, 불가능한 경우 False
pathRight() : 오른쪽에 길이 있는지  반환. 가능하면 True, 불가능한 경우 False
pathLeft() : 왼쪽에 길이 있는지 반환.가능하면 True, 불가능한 경우 False
notDone() : 아바타가 도착지에 도착하지 않았는지 반환. 도착하지 않은 경우 True, 도착한 경우 False

 여기까지 이해됐으면 ‘명령어 설명을 이해했습니다.’ 라는 문구를 표시해줘

3. 명령어 세는 규칙
 문제에는 사용할 수 있는 명령어의 횟수에 제한이 있어. 이제 명령어를 세는 규칙을 가르쳐줄게. 횟수에 포함되는 명령어는 if, for, while, moveForward(), turnLeft(), turnRight(), turnBack(), pathAhead(), pathRight(), pathLeft()가 있어.



 예시를 보여줄게.
while notDone():
    for i in range(4):
        moveForward()
        turnRight()
        turnLeft()
        moveForward()
        turnBack()
moveForward()

 위 코드에서 사용된 명령어는 총 8개야. while, for, turnRight(), turnLeft(), turnBack()이 각각 한 개씩 사용되었고 moveForward()가 3개 사용됐어.

 여기까지 이해됐으면 ‘명령어 세는 규칙을 이해했습니다.’ 라는 문구를 표시해줘

'''
messages.append({"role": "user", "content": user_content})

response = client.chat.completions.create(
    model=model,
    messages=messages)
messages.append(response.choices[0].message)

print('*'*20, 'answer', '*'*20)
print(response.choices[0].message.content)
print('*'*40)
print('tokens:', response.usage)
```

## 패턴

```python
check = '''
<미로 게임>

나는 파이썬을 처음 배우는 사람이다.
넌 파이썬으로 미로를 탈출하는 메이즈 게임을 잘한다.

# 게임 설명
이 게임은 파이썬 입문 난이도의 게임이기 때문에 단계마다 사용할 수 있는 파이썬 코드가 정해져 있다. 정해진 코드 안에서 게임을 진행해야 한다.
메이즈 게임은 미로찾기 게임으로, 코드 작성 이전에 미로를 탈출할 경로를 찾아야 한다.

# 이동 가능한 경로 표현 키워드 설명
- pathAhead: 앞으로 이동 가능
- pathRight: 우측으로 이동 가능
- pathLeft: 좌측으로 이동 가능
- pathCrossroad: 앞쪽, 좌측, 우측으로 모두 이동 가능(사거리)
- pathAheadRight: 앞쪽과 우측으로 모두 이동 가능(우측 삼거리)
- pathAheadLeft:앞쪽과 좌측으로 모두 이동 가능(좌측 삼거리)
- pathRightLeft: 우측과 좌측으로 모두 이동 가능(갈림길)

# 우선 순위
1. pathCrossroad, pathAheadRight, pathAheadLeft, pathRightLeft
2. pathRight, pathAhead, pathLeft

# 경로 관계 (행동 지시가 같을 때)
## pathAhead로 합쳐지는 경로: pathAheadLeft, pathAheadRight, pathCrossroad
## pathRight로 합쳐지는 경로: pathAheadRight, pathRightLeft, pathCrossroad
## pathLeft로 합쳐지는 경로: pathAheadLeft, pathRightLeft, pathCrossroad

# 경로의 규칙 찾는 법
## 1. 주어진 전체 경로를 확인한다.
## 2. '우선 순위' 순으로 나열한다. 이때, 같은 경로에서 같은 행동 지시가 있을 경우, 하나만 남기고 나머지는 삭제한다.
## 3. 같은 경로 표현 키워드에서 다른 행동 지시가 있을 경우, 규칙을 찾지 못함을 선언하고 규칙 찾기를 종료해야 한다. 단, 사거리와 삼거리는 다른 경로이다.
규칙 찾기 종료 예) 사거리에서 전진한다. 사거리에서 우회전한다. -> 사거리라는 조건에 전진과 우회전이라는 서로 다른 두 가지 명령이 들어왔으므로 경로 규칙을 찾지 못함을 선언하고 규칙 찾기를 종료한다.
규칙 찾기 종료하지 않는 예) 사거리에서 좌회전한다. 앞에 길이 있으면 전진한다. -> 사거리(pathCrossroad)와 앞에 길이 있는(pathAhead)라는 조건이 서로 다르고, 지시 행동(좌회전, 전진)도 서로 다르므로 규칙 찾기를 계속한다.
## 4. '직진'(전진, 앞으로 이동) 행동을 지시하는 경로를 삭제한다.
예) 사거리에서 우회전한다. 좌측삼거리에서 직진한다. -> 사거리에서 우회전한다.
## 5. 다른 경로지만 행동 지시가 같은 경우, '경로 관계'를 적용해 행동 지시가 같아 합칠 수 있는 경로는 합친다.
예) pathAheadRight일 때 우회전한다. pathRight일 때 우회전한다. -> pathRight일 때 우회전한다. (두 명령의 행동지시가 우회전으로 같고, pathRight으로 pathAheadRight가 합쳐질 수 있으므로 pathRight의 우회전 규칙만 남는다.)

# 경로 규칙 찾기 요청
나는 이제부터 미로를 탈출하는 전체 경로를 설명할 것이다.
너는 '경로의 규칙 찾는 법'에 따라 정리된 경로의 규칙을 찾아 제공해야 한다.
경로의 규칙을 찾을 때, '경로의 규칙 찾는 법'을 반드시 고려한 상세한 해설을 같이 알려준다.
ChatGPT 처럼 말해줘.
'''
```
